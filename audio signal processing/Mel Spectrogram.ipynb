```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import os
import numpy as np
```

# Loading audio files with Librosa

```python
path = '/Users/caomi/Documents/deep learning/audio signal processing/audio_files/'
audio_file1 = 'sax-player-in-the-world-170625.mp3'
```

```python
ipd.Audio(os.path.join(path,audio_file1))
```

```python
# load audio files with librosa
audio1, sr = librosa.load(os.path.join(path,audio_file1))
```

# Mel filter banks

```python
filter_banks = librosa.filters.mel(n_fft = 2048, sr = 22050, n_mels=20)
```

```python
filter_banks.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(filter_banks,
                        sr=sr,
                        x_axis='linear')
plt.colorbar(format='%+2.f')
plt.show()
```

# Extracting Mel Spectrogram

```python
mel_spectrogram = librosa.feature.melspectrogram(y=audio1, sr=sr, n_fft=2048, hop_length=512, n_mels=40)
```

```python
mel_spectrogram.shape
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(log_mel_spectrogram,
                        x_axis='time',
                        y_axis='mel',
                        sr=sr)
plt.colorbar(format='%+2.f')
plt.show()
```
