```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load audio file in the player
audio_path = '/Users/caomi/Documents/deep learning/audio signal processing/audio_files/high-chant-83888.mp3'
ipd.Audio(audio_path)
```

```python
#load audio file
signal , sr =librosa.load(audio_path ,sr=44100)
```

```python
#plot waveform
plt.figure(figsize=(18,8))
librosa.display.waveshow(signal , sr=sr, alpha=0.5)
plt.show()
```

```python
# derive spectrum using ft
ft=sp.fft.fft(signal)
magnitude=np.absolute(ft)
frequency=np.linspace(0,sr,len(magnitude))
```

```python
#plot spectrum
plt.figure(figsize=(18,8))
plt.plot(frequency[:5000],magnitude[:5000]) #magnitude spectrum
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```
