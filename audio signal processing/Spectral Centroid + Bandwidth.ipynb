```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import IPython.display as ipd
```

# Loading audio files

```python
audio_file1='/Users/caomi/Documents/deep learning/audio signal processing/audio_files/Percussion/bass drum/bass-drum__phrase_mezzo-piano_rhythm.mp3'
audio_file2 = '/Users/caomi/Documents/deep learning/audio signal processing/audio_files/sax-player-in-the-world-170625.mp3'
```

```python
ipd.Audio(audio_file2)
```

```python
# load audio files with librosa
audio1 , sr = librosa.load(audio_file1)
audio2 , sr = librosa.load(audio_file2)
```

# Spectral centroid with Librosa

```python
FRAME_SIZE = 1024
HOP_SIZE =512
```

```python
sc_audio1 = librosa.feature.spectral_centroid(y=audio1, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
sc_audio2 = librosa.feature.spectral_centroid(y=audio2, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
```

```python
sc_audio1.shape
```

# Visualize spectral centroid

```python
frames1 = range(len(sc_audio1))
t1 = librosa.frames_to_time(frames1)

frames2 = range(len(sc_audio2))
t2 = librosa.frames_to_time(frames2)
```

```python
plt.figure(figsize=(25, 10))

plt.plot(t1, sc_audio1, color='b')
plt.plot(t2, sc_audio2, color='r')

plt.show()
```

## Calculate bandwidth

```python
ban_audio1 = librosa.feature.spectral_bandwidth(y=audio1, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
ban_audio2 = librosa.feature.spectral_bandwidth(y=audio2, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)[0]
```

```python
ban_audio1.shape
```

## Visualise bandwidth

```python
plt.figure(figsize=(25, 10))

plt.plot(t1, ban_audio1, color='b')
plt.plot(t2, ban_audio2, color='r')

plt.show()
```
