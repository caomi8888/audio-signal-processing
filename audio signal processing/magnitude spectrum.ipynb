```python
def plot_magnitude_spectrum(signal, title, sr, f_ratio=1):
    ft=np.fft.fft(signal)
    magnitude_spectrum = np.abs(ft)

    #plot magnitude spectrum
    plt.figure(figsize=(18,5))

    frequency = np.linspace(0,sr,magnitude_spectrum.shape[0])
    num_frequency_bins = int(len(frequency) * f_ratio)
    plt.plot(frequency[:num_frequency_bins], magnitude_spectrum[:num_frequency_bins])
    plt.xlabel('frequency (Hz)')
    plt.title(title)

    plt.show()
    
```

```python
plot_magnitude_spectrum(violin_c4, 'violin', sr, 0.1)
```

```python
plot_magnitude_spectrum(piano_c4, 'piano', sr, 0.1)
```

```python
plot_magnitude_spectrum(noise, 'noise', sr, 0.1)
```
